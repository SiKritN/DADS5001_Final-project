```python
import pandas as pd
import plotly.graph_objects as go

# Original data
data = {
    'Category': ['แย่ ใช้ไม่ได้เลย','พอใช้ แก้เยอะ','ดี แก้นิดหน่อย','ดีมาก ใช้ได้เลย'],
    'Count': [8, 25, 54, 53]
}

df = pd.DataFrame(data)

# Create a new column for the broad categories
df['Broad Category'] = ['Negative', 'Negative', 'Positive', 'Positive']

fig = go.Figure()

# Plotting
fig.add_trace(
    go.Bar(
        x=[df['Broad Category'], df['Category']],
        y=df['Count'],
        text=df['Count'],
        marker_color = ['rgb(255, 102, 102)','rgb(255, 153, 153)','rgb(144, 238, 144)','rgb(0, 100, 0)'],
        name="Assessing Text Generated from Images scores",))

# Add line between Negative and Positive
fig.add_shape(type="line",
    x0=1.5, y0=0, x1=1.5, y1=max(df['Count']),
    line=dict(color="black",width=0.9)
)

# Adjust the bar spacing
fig.update_layout(title="Assessing Text Generated from Images scores",title_x=0.5)

# Customize axes and layout
fig.update_xaxes(title_text='Valuation', showgrid=False)
fig.update_yaxes(title_text='Count', showgrid=False,tickvals=[20, 40, 60], ticktext=[20, 40, 60])
fig.update_layout(plot_bgcolor='white')

fig.show()
```

```python
mean_values = df2.mean()
colors = [
    'rgb(0, 100, 0)' if mean > 3.5 else 
    'rgb(144, 238, 144)' if mean > 2.5 else 
    'rgb(255, 153, 153)' for mean in mean_values
]
fig = px.bar(mean_values,text='value')
fig.update_layout(plot_bgcolor='white',showlegend=False)
fig.update_traces(marker_color=colors)
fig.update_layout(title="Person Average Score",title_x=0.5)


fig.add_shape(type="line",
    x0=0, y0=2.5, x1=15, y1=2.5,
    line=dict(color="red",width=0.9)
)

fig.add_annotation(
    x=14,  
    y=2.7,  
    text="Ok Line",
    showarrow=False,
    font=dict(size=12, color="Green"),
)

fig.update_yaxes(title_text='Average Score', showgrid=False,range=[1, 5] ,tickvals=[1, 2, 3, 4], ticktext=['แย่ ใช้ไม่ได้เลย','พอใช้ แก้เยอะ','ดี แก้นิดหน่อย','ดีมาก ใช้ได้เลย'])
fig.update_xaxes(title_text='Person', showgrid=False,tickvals=list(range(1, 15)))

fig.show()
```
